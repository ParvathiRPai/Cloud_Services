```
def CreateSchemaToPandas(file_location):
  file_type = "csv"
  # CSV options
  infer_schema = "true"
  first_row_is_header = "true"
  delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
  df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)
  
  df_pandas=df.toPandas()
  return df_pandas
```

```
file_location_salaries = "/FileStore/tables/salaries.csv"
file_location_dept_manager = "/FileStore/tables/dept_manager.csv"
file_location_dept_emp = "/FileStore/tables/dept_emp.csv"
file_location_departments = "/FileStore/tables/departments.csv"
file_location_employees = "/FileStore/tables/employees.csv"
file_location_titles = "/FileStore/tables/titles.csv"
```

```
df_salaries=CreateSchemaToPandas(file_location_salaries)
display(df_salaries)
df_dept_manager=CreateSchemaToPandas(file_location_dept_manager)
display(df_dept_manager)
df_emp=CreateSchemaToPandas(file_location_dept_emp)
display(df_emp)
df_departments=CreateSchemaToPandas(file_location_departments)
display(df_departments)
df_employees=CreateSchemaToPandas(file_location_employees)
display(df_employees)
df_titles=CreateSchemaToPandas(file_location_titles)
display(df_titles)
```

```
import pandas as pd
import numpy as np
#  Type casting object type to string type
df_employees['hire_date']=pd.to_datetime(df_employees.hire_date)
df_employees['birth_date']=pd.to_datetime(df_employees.birth_date)
df_employees['age_when_hired'] = (df_employees['hire_date'] - df_employees['birth_date'])/np.timedelta64(1, 'Y')
df_employees['age_when_hired'] = df_employees.age_when_hired.astype(int)
```

```
grouped_salaries = df_salaries.groupby('emp_no')
df_employees['salary_dif'] = np.nan
for i, (name, group) in enumerate(grouped_salaries):
		salary_dif = max(group.salary) - min(group.salary)
		df_employees.loc[i, 'salary_dif'] = salary_dif
```

```
age_range = range(20, 48)
avg_salary = list(np.zeros(len(age_range)))
for i, age in enumerate(age_range):
	avg_salary[i] = np.mean(df_employees[df_employees['age_when_hired'] == age].salary_dif)
plt.plot(age_range, avg_salary)
plt.ylabel('Salary difference by age')
plt.xlabel('Age')
```
